```python
import tdcrpy as td
import matplotlib.pyplot as plt
import numpy as np
```

```python
radionuclide = "H-3"
mode = "beta-" # 'beta-' or 'beta+'
level = 'tot' # 0,1,2,3 .... or 'tot'
```

## Read information about the BetaShape calculataion

```python
td.TDCR_model_lib.readBetaShapeInfo(radionuclide,mode,level)
```

## Read the energy spectra tabulated from BetaShape Calculation

```python
energy, probability = td.TDCR_model_lib.readBetaShape(radionuclide,mode,level)

plt.figure("Energy spectrum")
plt.clf()
plt.plot(energy[:-1], probability)
plt.xlabel(r'$E$ /keV', fontsize=14)
plt.ylabel(r'$p(E)$ /(keV$^{-1}$)', fontsize=14)
```
